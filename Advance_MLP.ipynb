```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Define function and gradient
def f(x):
    return x ** 2

def grad_f(x):
    return 2 * x

# Gradient Descent Setup
epochs = 100
x_vals = [8]
lr = 0.1
for _ in range(epochs - 1):
    x_vals.append(x_vals[-1] - lr * grad_f(x_vals[-1]))

# Create figure
fig, ax = plt.subplots()
x_line = np.linspace(-10, 10, 400)
y_line = f(x_line)
ax.plot(x_line, y_line, label='$y = x^2$')

# Initialize moving point and text
point, = ax.plot([], [], 'ro', label='Current x')
text = ax.text(0.05, 0.9, '', transform=ax.transAxes)

# Plot settings
ax.set_xlim(-10, 10)
ax.set_ylim(0, 100)
ax.set_title('Gradient Descent on $y = x^2$')
ax.legend()

# Update function for animation
def update(frame):
    x = x_vals[frame]
    point.set_data([x], [f(x)])
    text.set_text(f'Epoch: {frame}, x = {x:.4f}')
    return point, text

# Create animation
ani = FuncAnimation(fig, update, frames=epochs, interval=200, blit=True)

# Display animation in Jupyter or Colab
HTML(ani.to_jshtml())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Hm f(x)
def f(x):
    return (x**2 * np.sin(x) + 6*x)*0.1

# Sai phn trung tm
def central_diff(f, h=1e-6):
    def df(x):
        return (f(x + h) - f(x - h)) / (2 * h)
    return df

df = central_diff(f)

# Tham s
epochs = 100
x_vals = [7.5]
initial_lr = 0.01

# Adaptive learning rate decay
def adaptive_lr(epoch, decay_rate=0.5, decay_step=25):
    return initial_lr * (decay_rate ** (epoch // decay_step))

# Adam optimizer with bias correction
m, v, t = 0, 0, 0
def Adam(x, lr=0.5, beta1=0.9, beta2=0.4):
    global m, v, t
    t += 1
    g = df(x)
    m = beta1 * m + (1 - beta1) * g
    v = beta2 * v + (1 - beta2) * (g ** 2)
    m_hat = m / (1 - beta1 ** t)
    v_hat = v / (1 - beta2 ** t)
    return x - lr * m_hat / (np.sqrt(v_hat) + 1e-8)

# Qu trnh hc
for epoch in range(epochs - 1):
    x_prev = x_vals[-1]
    x_next = Adam(x_prev)
    x_vals.append(x_next)

# D liu v
x_line = np.linspace(-10, 10, 400)
y_line = f(x_line)

#  th
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x_line, y_line, label='$f(x) = x^2 \\sin(x)$', color='blue')
point, = ax.plot([], [], 'ro', label='Current x')
text = ax.text(0.05, 0.9, '', transform=ax.transAxes)

ax.set_title('Gradient Descent vi Adam Optimizer')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend()
ax.grid(True)

# Hm cp nht cho animation
def update(frame):
    x = x_vals[frame]
    y = f(x)
    point.set_data([x], [y])
    lr_disp = adaptive_lr(frame)
    text.set_text(f'Epoch: {frame}, x = {x:.4f}, lr = {lr_disp:.4f}')
    return point, text

# To animation
ani = FuncAnimation(fig, update, frames=epochs, interval=100, blit=True)
HTML(ani.to_jshtml())
```

```python
# --- (Norm) ---

def Normalize_guassian(data):
    mean = np.mean(data, axis = 0)
    std = np.std(data, axis=0)
    return (data - mean) / (std + 1e-8)

X = Normalize_guassian(X)
X_test = Normalize_guassian(X_test)

# --- One-hot encoding ---
Y = np.array(pd.get_dummies(Y))
Y_test = np.array(pd.get_dummies(Y_test))
```

# Activation Function & o hm


```python
# --- Hm kch hot (Activation functions) ---
def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))

def relu(Z):
    return np.maximum(0, Z)

def leaky_relu(Z):
    return np.where(Z > 0, Z, Z * 0.01)

def leaky_relu_back(dl, Z, alpha=0.01):
    dx = np.ones_like(Z)
    dx[Z < 0] = alpha
    return dx * dl

def sigmoid_back(dA, Z):
    return dA * sigmoid(Z) * (1 - sigmoid(Z))

def relu_back(dA, Z):
    dZ = np.array(dA, copy=True)
    dZ[Z <= 0] = 0
    return dZ

def softmax(u):
    u_exp = np.exp(u - np.max(u, axis=1, keepdims=True))
    return u_exp / np.sum(u_exp, axis=-1, keepdims=True)
```

```python
# nh ngha cu trc layer_dim: [s n-ron u vo, s n-ron u ra]
# E.g: layer_dim = [[784, 256], [256, 256], [256,128],[128,64], [64, 10]]
def init_weight_bias(layer_dim):
    W = []
    for i in layer_dim:
        weight = np.random.randn(i[0] + 1, i[1]) * np.sqrt(2 / i[1])
        # He Ini
        # i[0] + 1 ,i[1] s input node + 1 (bias), output node
        W.append(weight)
    return W

```

## 2. Tch batch train

```python
# --- = DataLoader nu dng th vin ---
def next_batch(X, Y, batch_size):
    for i in range(0, X.shape[0], batch_size):
        yield X[i:i + batch_size], Y[i:i + batch_size]
```

## 2.Forward pass


```python
class Forward_layers:
    def __init__(self, layer):
        self.W = layer["weights"]
        self.activation = layer["activations"]
        self.layer_dim = layer["layer_dim"]
        self.num_layer = len(self.layer_dim)

    def fit(self, A_prev, W, activation):
        A_prev_bias = np.hstack((A_prev, np.ones((A_prev.shape[0], 1))))
        Z_curr = np.dot(A_prev_bias, W)

        if activation == "relu":
            return relu(Z_curr), Z_curr
        elif activation == "sigmoid":
            return sigmoid(Z_curr), Z_curr
        elif activation == "leaky_relu":
            return leaky_relu(Z_curr), Z_curr
        elif activation == "softmax":
            return softmax(Z_curr), Z_curr
        raise ValueError("Invalid act. func")

    def forward(self, X):
        memory_A, memory_Z = [], []
        A_curr = X
        for i in range(self.num_layer):
            A_prev = A_curr
            A_curr, Z_curr = self.fit(A_prev, self.W[i], self.activation[i])
            memory_A.append(A_prev)
            memory_Z.append(Z_curr)
        return A_curr, Z_curr, memory_A, memory_Z
```

## 3.Backward pass

```python
class Backward_layers:
    def __init__(self, layer):
        self.activation = layer["activations"]
        self.layer_dim = layer["layer_dim"]
        self.num_layer = len(self.layer_dim)

    def fit(self, dA_curr, W_curr, Z_curr, A_prev, activation_back):
        N = A_prev.shape[0]

        A_prev_bias = np.hstack((A_prev, np.ones((N, 1))))

        if activation_back == "relu":
            dZ_curr = relu_back(dA_curr, Z_curr)
        elif activation_back == "sigmoid":
            dZ_curr = sigmoid_back(dA_curr, Z_curr)
        elif activation_back == "leaky_relu":
            dZ_curr = leaky_relu_back(dA_curr, Z_curr)
        else:
            dZ_curr = dA_curr


        dW_curr = np.dot(A_prev_bias.T, dZ_curr) / N

        # Ct cui cng ca dA_prev_with_bias l gradient i vi bias, khng cn truyn ngc, Nn chng ta ch ly cc ct u tin (tng ng vi cc feature)
        dA_prev = np.dot(dZ_curr, W_curr[:-1, :].T)

        return dA_prev, dW_curr

    def backward(self, Y_h, Y, memory_W, memory_A, memory_Z):
        dA_prev = Y_h - Y
        gradient_W = []
        for i in range(self.num_layer):
            dA_curr = dA_prev
            W_curr = memory_W[-i - 1]
            Z_curr = memory_Z[-i - 1]
            A_prev = memory_A[-i - 1]

            dA_prev, dW_curr = self.fit(dA_curr, W_curr, Z_curr, A_prev, self.activation[-i - 1])
            gradient_W.append(dW_curr)

        return gradient_W
```

## 4. T build 1 class Sequential
1 class k tha t 2 class trn v to ra 1 model c th chy forward ln backward, v cung cp nhng hm cn thit   tnh loss, accuracy

```python
import numpy as np

# Assuming these classes and functions are defined elsewhere in your code:
# Forward_layers, Backward_layers, softmax

class ANN(Forward_layers, Backward_layers):
    def __init__(self, layer, clip_gradient_value=None, lr=0.001, beta1=0.9, beta2=0.999, epsilon=1e-8):
        super().__init__(layer)
        self.layer = layer
        self.clip_gradient_value = clip_gradient_value
        self.lr = lr  # Learning rate for Adam
        self.beta1 = beta1
        self.beta2 = beta2
        self.epsilon = epsilon
        self.t = 0  # Time step for Adam

        # Initialize Adam's m and v for each combined weight/bias array
        # self.layer["weights"] is now a list where each element is a combined W_i, b_i
        self.m_params = [np.zeros_like(param_array) for param_array in self.layer["weights"]]
        self.v_params = [np.zeros_like(param_array) for param_array in self.layer["weights"]]


    def fit(self, X, Y):
        self.t += 1 # Increment time step

        forward = Forward_layers(self.layer)
        A_curr, Z_curr, memory_A, memory_Z = forward.forward(X)

        Y_h = A_curr
        backward = Backward_layers(self.layer)

        # Now, backward.backward should return a single list of gradients for the combined parameters
        # Assuming gradient_params is a list where gradient_params[i] corresponds to self.layer["weights"][-i-1]
        gradient_params = backward.backward(Y_h, Y, self.layer["weights"], memory_A, memory_Z)

        # Apply Adam optimization
        # Iterate through the layers using self.layer_dim as before
        for i in range(len(self.layer_dim)):
            # Use the same indexing logic you had: -i-1 for parameters, i for gradients
            param_index = -i - 1
            current_gradient = gradient_params[i]

            # Adam update for combined parameters (weights and biases)
            self.m_params[param_index] = self.beta1 * self.m_params[param_index] + (1 - self.beta1) * current_gradient
            self.v_params[param_index] = self.beta2 * self.v_params[param_index] + (1 - self.beta2) * (current_gradient ** 2)

            m_hat = self.m_params[param_index] / (1 - self.beta1 ** self.t)
            v_hat = self.v_params[param_index] / (1 - self.beta2 ** self.t)

            self.layer["weights"][param_index] -= self.lr * m_hat / (np.sqrt(v_hat) + self.epsilon)

    def forward(self, X):
        forward = Forward_layers(self.layer)
        A_curr, _, _, _ = forward.forward(X)
        return A_curr

    def predict(self, X):
        Y_h_prob = self.forward(X)
        return np.where(softmax(Y_h_prob) >= np.max(softmax(Y_h_prob), axis=1, keepdims=True), 1, 0)

    def predict_prob(self, X):
        return self.forward(X)

    def score(self, X, Y):
        Y_h_prob = self.forward(X)
        Y_h_class = np.where(softmax(Y_h_prob) >= np.max(softmax(Y_h_prob), axis=1, keepdims=True), 1, 0)
        return np.mean([np.array_equal(Y_h_class[i], Y[i]) for i in range(len(Y))]) * 100 # Accuracy thang 100

    def cost(self, X, Y):
        Y_h_prob = self.forward(X)
        return -np.mean(np.sum(Y * np.log(softmax(Y_h_prob) + 1e-10), axis=1))
```

# Train

```python
# --- Setup ---
batch_size = 256
epochs = 400
lr = 0.01
layer_dim = [[784, 256], [256, 10]]
memory_W = init_weight_bias(layer_dim)
layer = {"layer_dim": layer_dim, "activations": [ "relu","softmax"], "weights": memory_W}


# Training loop ---
model = ANN(layer)
losses_train, losses_test, accuracies_train, accuracies_test = [], [], [], []
for epoch in range(epochs):
    indices = np.arange(X.shape[0])
    np.random.shuffle(indices)
    X_train = X[indices]
    y_train = Y[indices]
    loss_train, acc_train = [], []
    for batch_X, batch_Y in next_batch(X_train, y_train, batch_size):
        model.fit(batch_X, batch_Y)
        loss_train.append(model.cost(batch_X, batch_Y))
        acc_train.append(model.score(batch_X, batch_Y))
    losses_train.append(np.mean(loss_train))
    accuracies_train.append(np.mean(acc_train))
    losses_test.append(model.cost(X_test, Y_test))
    accuracies_test.append(model.score(X_test, Y_test))

    print(f"Epoch {epoch+1}, Train Loss: {np.mean(loss_train):.4f}, Train Acc: {np.mean(acc_train):.2f}%", f"Test Loss: {losses_test[-1]:.4f}, Test Acc: {accuracies_test[-1]:.2f}%")
```
